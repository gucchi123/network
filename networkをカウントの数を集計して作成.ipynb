```python
import requests
import json

items = []
params = {"page":1, "per_page":100}
for i in range(60):
    print("fetching... page " + str(i+1))
    params["page"] = i + 1
    res = requests.get("https://qiita.com/api/v2/items", params=params)
    items.extend(json.loads(res.text))
```

```python
items[-8]
```

```python
tags_list = []
for item in items[:-9]:
    tags = [tag["name"] for tag in item["tags"]]
    tags_list.append(tags)
```

```python
tags_list
```

```python
import collections
import itertools

tag_count = collections.Counter(itertools.chain.from_iterable(tags_list)).most_common(50)
```

```python
tag_count
```

```python
import networkx as nx
G = nx.Graph()
G.add_nodes_from([(tag, {"count":count}) for tag, count in tag_count])
```

```python
for tag, count in tag_count[:20]:
    print(count)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
nx.draw_networkx(G)
```

```python
for tag in tags_list:
    #print(tag)
    for node0, node1 in itertools.combinations(tag,2):
        print(node0, node1)
        if not G.has_node(node0) or not G.has_node(node1):
            continue
        if G.has_edge(node0, node1):
            G.adj[node0][node1]["weight"] += 1
        else:
            G.add_edge(node0, node1, weight=1)
```

```python
plt.figure(figsize=(15, 15))

pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)

plt.axis("off")
plt.savefig("default.png")
plt.show()
```

```python
#エッジの刈り取り
#エラーが出ないようにhttp://stronger.hatenablog.com/entry/2016/12/13/141028によりlist化
for (u, v, d) in list(G.edges(data=True)):
    if d["weight"] <= 4:
        print(d)
        G.remove_edge(u, v)
```

```python
pos = nx.spring_layout(G, k=0.3)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math


plt.figure(figsize=(15,15))
pos = nx.spring_layout(G, k=0.5)

node_size = [ d['count']*20 for (n,d) in G.nodes(data=True)]
nx.draw_networkx_nodes(G, pos, node_color='r',alpha=0.6, node_size=node_size)
nx.draw_networkx_labels(G, pos, fontsize=14, font_family="Yu Gothic", font_weight="bold")

edge_width = [ d['weight']*0.2 for (u,v,d) in G.edges(data=True)]
nx.draw_networkx_edges(G, pos, alpha=0.4, edge_color='b', width=edge_width)

plt.axis('off')
plt.savefig("g2.png")
plt.show()
plt.tight_layout()
```
